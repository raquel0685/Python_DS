RESPONDENDO AS NOVAS PERGUNTAS DO CEO:

```python
import pandas as pd

dados = pd.read_csv('AB_NYC_2019.csv')

dados.head()
```

Categorias de imóveis que estão cadastradas
dentro da base de dados da cidade de Nova York:

- Selecionar coluna 'room_typ';
- Realizar contagem única dos valores

```python
room_type = dados.loc[:, 'room_type']

import numpy as np

np.unique(room_type)
```

Quantidade usuários (Hosts) únicos cadastrados existem
dentro da base de dados da cidade de Nova York:

- Selecionar a coluna 'host_id';
- Realizar contagem única dos valores;

```python
#selecao de linhas e colunas
host_id = dados.loc[:,'host_id']

#separacao dos valores únicos
host_unique = np.unique(host_id)

#contagem dos valroes únicos
len(host_unique)
```

Variação do preços dos imóveis em NY:

- Selecao de linhas e colunas;

- Aplicar operador matemárico do dp;

```python
precos = dados.loc[:,'price']

print('Média: {:.2f}'.format(np.mean(precos)))

print('Desvio padrão: {:.2f}'.format(np.std(precos)))
```

```python
np.min(precos)
```

São imóveis baratos ou caros:

- Desenhar um histograma para mostrar o número de apartamentos dentro
da faixa do valor do aluguel.

```python
linhas = dados.loc[:,'price'] < 1100

precos = dados.loc[linhas,'price']

from matplotlib import pyplot as plt

#maior aluguel
print(np.max(precos))

plt.hist(precos,bins=11);
```

Qual a distribuição do número de Reviews? Existem
imóveis com muitos e outro com poucos reviews?


```python
linhas = dados.loc[:,'number_of_reviews'] < 300

reviews = dados.loc[linhas,'number_of_reviews']

from matplotlib import pyplot as plt

#maior aluguel
print(np.max(reviews))

#bins: 300-0 (maior - menor) / 10 (intervalo)
plt.hist(reviews, bins=30);
```
