Qual é o valor do aluguel (diária) mais caro de cada região da base de dados da cidade de Nova York?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import folium

data = pd.read_csv('AB_NYC_2019.csv')
data.head(2)
```

```python
#seleção linhas/colunas, agrupar e aplicar operador de máximo

colunas = ['price','neighbourhood_group']
colunas_groupby =['neighbourhood_group']
#reset_index: transforma o nome das linhas em números do index, e deixa os nomes da busca como nomes de coluna
data_plot = data.loc[:,colunas].groupby(colunas_groupby).max().reset_index()
#data_plot = data.loc[:,colunas].groupby(colunas_groupby).mean().reset_index()

#gráfico

px.bar(data_plot, x='neighbourhood_group',y='price')
```

Conseguimos saber onde estão localizados os imóveis com o valor do aluguel mais caro, na cidade de Nova York?

```python
colunas = ['price','neighbourhood_group','latitude','longitude']
colunas_groupby =['neighbourhood_group']

data_plot = data.loc[:,colunas].groupby(colunas_groupby).max().reset_index()
```

```python
f = folium.Figure(width=1024, height=768)
map = folium.Map(location=[data_plot['latitude'].mean(),
                     data_plot['longitude'].mean()],
            zoom_start=14,
           control_scale=True)

for index, location_info in data_plot.iterrows():
  folium.Marker([location_info['latitude'],location_info['longitude']],popup=location_info['neighbourhood_group']).add_to(map)
map
```

Conseguimos saber onde estão localizados os imóveis pelo seu tipo?

```python
#conjuto de dados (fitragem linhas/colunas)
colunas = ['neighbourhood_group','room_type','latitude','longitude']
data_plot = data.loc[:,colunas].sample(20)
data_plot

#criar uma nova coluna chamada 'color'
data_plot.loc[:,'color'] = 'NA'
data_plot

linhas_private_room = data_plot.loc[:,'room_type'] == 'Private room'
linhas_shared_room = data_plot.loc[:,'room_type'] == 'Shared room'
linhas_entire_home = data_plot.loc[:,'room_type'] == 'Entire home/apt'

data_plot.loc[linhas_private_room,'color'] ='darkgreen'
data_plot.loc[linhas_shared_room,'color'] ='darkred'
data_plot.loc[linhas_entire_home,'color'] ='darkblue'
data_plot

#desenha mapa
#add pins
map = folium.Map()

for index, location_info in data_plot.iterrows():
  folium.Marker([location_info['latitude'],location_info['longitude']],
                popup=location_info[['neighbourhood_group','room_type']],
                icon=folium.Icon(color=location_info['color'])).add_to(map)
map

```
