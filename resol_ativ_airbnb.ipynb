```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('airbnb_ny_data.csv')
data.head()
```

Qual é o intervalo de variação do preço dos alugueis dos imóveis?

```python
#intervalo de variação = dp ou variancia:

coluna = 'price'
data.loc[:,coluna].std()

#outro jeito de fazer:
media = data.loc[:,coluna].mean()
desvio_padrao = np.std(data.loc[:,coluna])

print('O preço médio é US${:.2f} +/- US${:.2f}.'.format(media,desvio_padrao))
```

Qual a distribuição dos imóveis de acordo com o número mínimo de
pernoites que podem ser agendado?





```python
coluna = 'minimum_nights'
linhas = data.loc[:,coluna]<100

data_plot = data.loc[linhas,coluna]

#histograma
#bins: valor maior - valor menor / numero de colunas desejada

plt.hist(data_plot, bins=20);
```

Qual a média e a mediana de imóveis cadastrado por host (dono do
imóvel)?

```python
coluna = 'calculated_host_listings_count'

media = data.loc[:,coluna].mean()
mediana = np.median(data.loc[:,coluna])

print('A média de imóveis cadastrados é {:.2f} e a mediana é {:.2f}.'.format(media,mediana))
```

Como você interpretaria o seguinte histograma?


```python
colunas = 'reviews_per_month'
linhas = data.loc[:, 'reviews_per_month'] < 10
data_plot = data.loc[linhas, colunas]
plt.hist( data_plot, bins=10 );
plt.title( "Numero de Avaliações por Mês");
plt.xlabel( 'Quantidade de Avaliações');
plt.ylabel( 'Quantidade de Imóveis');
```

###### Muitos imóveis têm até 1 avaliação, 4x mais do que aqueles que possuem até 2 avaliações. Pouqíssimos imóveis tem muitas avaliações.

Qual a data mais recente da última avaliação

```python
coluna = 'last_review'

avaliacoes = data.loc[:,coluna].sort_values(ascending=True)
data_recente = avaliacoes.loc[0]
print('A data mais recente das avaliações é {}.'.format(data_recente))
```

Qual o valor máximo de pernoites agendadas?

```python
coluna = 'minimum_nights'

maximo_pernoite = data.loc[:,coluna].max()

#data.loc[:,coluna].sort_values(ascending=False)
print('O número máximo de pernoites é {}.'.format(maximo_pernoite))
```

Como você interpretaria o seguinte valor do desvio padrão?

```python
print( 'Preço medio da concorrência é de U${} +/- U${}'.format( 180, 20 ) )
```

###### DP é a dist média dos pontos em torno do onto central.

Quantos nomes de donos de imóveis são únicos?

```python
coluna = 'host_id'
hosts = data.loc[:,coluna]
hosts_unicos = len(np.unique(hosts))

print('O núemro de hosts é {}.'.format(hosts_unicos))
```

Como você interpretaria o seguinte histograma:

```python
colunas = 'latitude'
data_plot = data.loc[:, colunas]
plt.hist( data_plot, bins=16 );
plt.title( "Posição geográfica dos imóveis")
plt.xlabel( 'Valor da Latitude');
plt.ylabel( 'Quantidade de Imóveis');
```

###### A maioria dos imóveis está concentrado/centralizado numa mesma região (latitude 40.7).

Quantos identificadores únicos existem na base de dados?

```python
coluna = 'id'

id = data.loc[:,coluna]
id_unico = len(np.unique(id))

print('O número de ids únicos é {}.'.format(id_unico))
```
